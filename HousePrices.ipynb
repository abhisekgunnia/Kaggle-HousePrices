```python
#Home Price - Applying my thought

#Y = f(x)

#Y - Prediction target. i.e., Dependent Variable
#X - Features. i.e., Independent Variables
#f - Model -> This model will have to be opmitimized to predict Y with acceptable accuracy. This Optimization has to happen programmatically, so that it meets the definition of #Machine Learning Algorithm.

#A computer program is said to learn from experience E with respect to some class of tasks T and performance measure P if its performance at tasks in T, as measured by P, #improves with experience E.

#Apply this definition to our context.

#T -> Task the program will learn from. i.e., X
#P -> Accuracy of Y
#E -> f

Y = home_data.SalePrice
X = home_data[['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']]

print('Sample Prediction Traget : \n',Y.head())
print('Respective Features for above targets :\n',X.head())
```

```python
# We will use DecisionTreeRegressor Model from scikit Learn package
# Important the required modules
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
# Create a Model with random_state=1 => Read more about this
home_model = DecisionTreeRegressor(random_state=1)
# Fitting our training data into the model. This step is also call training the model.
home_model.fit(X,Y)
# Making predictions - This is also called testing the model
home_predictions = home_model.predict(X)
print(home_predictions)
print('Mean Absolute Error-MAE : ',mean_absolute_error(Y,home_predictions))

# Here the MAE is so low, and the model appears to be perfect. But the predicted values are from the same set of features that was used for training the model.
```

```python
# In the below approach, we will exclude data for validation.
from sklearn.model_selection import train_test_split
# Read more about this randome state later
train_X,val_X,train_Y,val_Y = train_test_split(X,Y,random_state=1)
home_model.fit(train_X,train_Y)
home_predictions_new = home_model.predict(val_X)
print(home_predictions_new)
print('Mean Absolute Error-MAE : ', mean_absolute_error(home_predictions_new,val_Y))

# Now we see a huge jump in MAE when we tested our model on a sample that was not included in training. A model with such a behaviour is not optimized and hence cannot be used for practical purposes.
```

```python
# Defining a function to calculate MAE

def getMae(no_of_leafs,train_X,val_X,train_Y,val_Y):
    model = DecisionTreeRegressor(max_leaf_nodes=no_of_leafs,random_state = 1)
    model.fit(train_X,train_Y)
    predictions = model.predict(val_X)
    outMae = mean_absolute_error(predictions,val_Y)
    return outMae
```

```python
getMae(150,train_X,val_X,train_Y,val_Y)
```

```python
for i in [10,20,40,60,90,100,150,200,300,400,500,700,900,1100,1500,2500,5000,10000]:
    print('MAE for ',i,'leafs is : ',getMae(i,train_X,val_X,train_Y,val_Y))
    
```
